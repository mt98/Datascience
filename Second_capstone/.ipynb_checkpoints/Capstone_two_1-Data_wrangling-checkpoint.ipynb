We will now look at the clinical data for the training dataset.

```python
Training_clinical=pd.read_csv('training_data/train_clinical_data.csv')
Training_clinical.head()
```

Let us have a look at the whether there are any missing values and what percentage of missing values there are for each attribute.

```python
Training_clinical.isnull().sum()
missing = pd.concat([Training_clinical.isnull().sum(), 100 * Training_clinical.isnull().mean()], axis=1)

missing.columns=["count", "%"]
missing.sort_values(by=["count", "%"])
```

**We know that there are missing values for our training clinical dataset. Very high percentage of missing values for updrs_4 and upd23b_clinical_state_on_medication. The upd23b_clinical_state_on_medication is Whether or not the patient was taking medication such as Levodopa during the UPDRS assessment.  UPDRS scores  we will impute the missing values .

What are the range of values for UPDRS_4 scores


```python
_=plt.hist(Training_clinical.updrs_4)
_=plt.ylabel("UPDRS_4 score")
_=plt.title("Distribution of UPDRS_4 scores")
plt.show()
```

**Since the most common value (mode) for UPDRS_4 score is 0, we will impute missing values of UPDRS_4 with it.The median is also 0 for this dataset.

```python
Training_clinical["updrs_4"] = Training_clinical["updrs_4"].fillna(0)
```

**For the rest of the scores we will will also impute missing values with median which is most robust to outliers

```python
Training_clinical["updrs_1"] = Training_clinical["updrs_1"].fillna(Training_clinical['updrs_1'].median())
Training_clinical["updrs_2"] = Training_clinical["updrs_2"].fillna(Training_clinical['updrs_2'].median())
Training_clinical["updrs_3"] = Training_clinical["updrs_3"].fillna(Training_clinical['updrs_3'].median())
```

Let us look at the range of visit_months in the training clinical dataset

```python

Training_clinical.groupby("visit_month").size()
```

**We can see that each patient did not show up to each visit every time. Is there a correlation between the UPDRS scores and the frequency of visits.
Let us look at how min difference between visit months is associated with UPDRS scores

```python
Training_clinical['visit_month_diff'] = Training_clinical.groupby(['patient_id'])['visit_month'].diff()
Training_clinical['visit_month_diff_min'] = Training_clinical.groupby('patient_id')['visit_month_diff'].transform('min')
Training_clinical['median_updrs_1'] = Training_clinical.groupby('visit_month_diff_min')['updrs_1'].transform('median')
Training_clinical['median_updrs_2'] = Training_clinical.groupby('visit_month_diff_min')['updrs_2'].transform('median')
Training_clinical['median_updrs_3'] = Training_clinical.groupby('visit_month_diff_min')['updrs_3'].transform('median')
Training_clinical['median_updrs_4'] = Training_clinical.groupby('visit_month_diff_min')['updrs_4'].transform('median')
```

```python
Month_UPDRS=Training_clinical[['visit_month_diff_min','median_updrs_1','median_updrs_2','median_updrs_3','median_updrs_4']]

t=sns.lineplot(x='visit_month_diff_min', y='value', hue='variable', 
             data=pd.melt(Month_UPDRS, ['visit_month_diff_min']))
_=t.set_xticks([3,6,12])
_=t.set(title='Relationship of UPDRS scores with visit frequency')
_=plt.ylabel("UPDRS_scores")
_=plt.xlabel("Minimum Visit frequency")
plt.show()
```

Visit frequency of patients has an association with the UPDRS scores. Median UPDRS 1 scores are lower for those patients which had a minimum visit frequency of 12 months. For UPDRS scores 2 and UPDRS scores 3 there is a decrease in median scores from 3 to 6 to 12 months. There is no association of UPDRS scores 4 with minimum visit frequency but again that may be due to many missing variables, which were imputed.
